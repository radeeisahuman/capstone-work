```python
import pandas as pd
from pathlib import Path

current_dir = Path.cwd()
root_dir = current_dir.parents[0]
data_dir = root_dir / 'data'

# read scraped data

data = pd.read_csv(data_dir / 'reed_course_reviews.csv')
print(data)
```

```python
data = data.dropna(axis='rows', how='all', subset='review_text')
print(data)
```
