```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

extracted_dict = {
    'course_title': [],
    'course_link': []
}

def extract(page, page_num):
    print(f'Scraping page: {page_num}')
    response = requests.get(page)
    soup = BeautifulSoup(response.content, 'html.parser')
    courses = soup.find_all('article', class_='course-card')
    for course in courses:
        course_title = course.find('h2', class_='course-card-title')
        course_link = course_title.find('a')['href']
        extracted_dict['course_title'].append(course_title.text.strip())
        extracted_dict['course_link'].append(course_link)

for i in range(1,101):
    extract(f'https://www.reed.co.uk/courses/discount?pageno={i}', i)
```

```python
course_links = pd.DataFrame(extracted_dict)
course_links['course_link'] = 'https://www.reed.co.uk' + course_links['course_link']
print(course_links.head())
```

```python
course_links.to_csv('cleaned_course_links.csv', index=False)
```
